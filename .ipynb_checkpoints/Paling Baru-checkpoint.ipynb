Laboratory Data:

- TCHOL_L: Cholesterol - Total

Questionnaire Data:

- MCQ_L: Medical Conditions

- SMQ_L: Smoking - Cigarette Use

```python
columns_to_filter = {
    # katagori
    'diabetes': [3],
    'alkohol': [99, 77],
    'sedih-depresi-putus_asa': [9, 7],
    'gangguan_tidur': [9, 7],
    'gangguan_makan': [9, 7],
    'riw_liver': [9, 7],
    'riw_tiroid': [9, 7],
    'riw_kanker': [9, 7],
    'riw_kolesterol_tinggi': [9, 7],
    'merokok100': [9, 7]
}
for col, values in columns_to_filter.items():
    if pd.api.types.is_numeric_dtype(df[col]):
        df = df.loc[~df[col].isin(values)].reset_index(drop=True)
    else:
        df = df.loc[~df[col].isin(values)].reset_index(drop=True)
        df[col] = df[col].cat.remove_unused_categories()
```

### 3.2 Membuat Fitur Tekanan Darah

**Mean Arterial Pressure**</br>
MAP dapat dihitung menggunakan rumus berikut:</br>
Diastolik + 1/3 (Sistolik - Diastolik)

```python
rata2_sistolik = (df['sistolik1'] + df['sistolik2'] + df['sistolik3'])/3
rata2_diastolik = (df['diastolik1'] + df['diastolik2'] + df['diastolik3'])/3
df['tekanan_darah'] = rata2_diastolik + (rata2_sistolik - rata2_diastolik)/3
df['tekanan_darah'] = df['tekanan_darah'].round().astype('int64')
```

# 6. Feature Importance

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from sklearn.base import clone
import numpy as np
import pandas as pd

def cross_validated_permutation_importance(model, X, y, metric=roc_auc_score, cv=5, random_state=42):
    feature_importances = {col: [] for col in X.columns}
    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=random_state)

    for train_idx, val_idx in skf.split(X, y):
        X_train_fold, X_val_fold = X.iloc[train_idx], X.iloc[val_idx]
        y_train_fold, y_val_fold = y.iloc[train_idx], y.iloc[val_idx]

        # Clone & train model untuk fold ini
        model_fold = clone(model)
        model_fold.fit(X_train_fold, y_train_fold)

        # Skor baseline
        y_pred = model_fold.predict_proba(X_val_fold)[:, 1]
        baseline = metric(y_val_fold, y_pred)

        for col in X.columns:
            X_val_permuted = X_val_fold.copy()
            X_val_permuted[col] = np.random.permutation(X_val_permuted[col].values)
            y_pred_perm = model_fold.predict_proba(X_val_permuted)[:, 1]
            importance = baseline - metric(y_val_fold, y_pred_perm)
            feature_importances[col].append(importance)

    result = pd.DataFrame({
        'feature': list(feature_importances.keys()),
        'mean_importance': [np.mean(imp) for imp in feature_importances.values()],
        'std_importance': [np.std(imp) for imp in feature_importances.values()]
    }).sort_values(by='mean_importance', ascending=False).reset_index(drop=True)

    return result
```

```python
# Untuk model Random Forest (misal clf_rf sudah dari RandomizedSearchCV)
model_rf = clf_rf.best_estimator_ if hasattr(clf_rf, "best_estimator_") else clf_rf
cvpfi_rf = cross_validated_permutation_importance(model_rf, X_train, y_train)

model_xgb = clf_xgb.best_estimator_ if hasattr(clf_xgb, "best_estimator_") else clf_xgb
cvpfi_xgb = cross_validated_permutation_importance(model_xgb, X_train, y_train)

# Untuk SVM
model_svm = clf_svm.best_estimator_ if hasattr(clf_svm, "best_estimator_") else clf_svm
cvpfi_svm = cross_validated_permutation_importance(model_svm, X_train, y_train)

# Untuk KNN
model_knn = clf_knn.best_estimator_ if hasattr(clf_knn, "best_estimator_") else clf_knn
cvpfi_knn = cross_validated_permutation_importance(model_knn, X_train, y_train)
```
